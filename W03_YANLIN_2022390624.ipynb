```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
data.info()

sns.countplot(x='EnvironmentSatisfaction', data=data)
plt.title('Distribution of Environment Satisfaction Levels')
plt.show()

# 计算不同满意度水平的流失率
satisfaction_attrition = data.groupby('EnvironmentSatisfaction')['Attrition'].value_counts(normalize=True).unstack()
satisfaction_attrition.plot(kind='bar', stacked=True)
plt.title('Attrition Rate by Environment Satisfaction')
plt.ylabel('Percentage')
plt.show()
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

data = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
data.info()
# 创建交叉表
contingency_table = pd.crosstab(data['EnvironmentSatisfaction'], data['Attrition'])
chi2, p, _, _ = chi2_contingency(contingency_table)

print(f'Chi-Square Test Statistic: {chi2}, p-value: {p}')
```
